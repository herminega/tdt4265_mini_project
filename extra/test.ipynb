```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# Load a mask file (change path to an actual mask)
mask_path = "/cluster/projects/vc/data/mic/open/HNTS-MRG/train/10/preRT/10_preRT_mask.nii.gz"
mask = nib.load(mask_path).get_fdata()

# Plot a central slice to inspect the labels
plt.figure(figsize=(8, 8))
plt.imshow(mask[:, :, mask.shape[2] // 2], cmap="gray")  # Show middle slice
plt.colorbar()
plt.title("Example Segmentation Mask")
plt.show()

# Print unique values
print("Unique values in mask:", np.unique(mask))
```
